```python
import pandas as pd
import numpy as np

datos = pd.read_csv('AMZN.csv')

datos.head()
```

```python
datos['Close'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(datos['Date'], datos['Close'], label='Precio de Cierre', color='blue')
plt.plot(datos['Date'], datos['Open'], label='Precio de Inicio', color='red')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.title('Historia de Precios de Cierre de Amazon')
plt.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

datos['Date'] = pd.to_datetime(datos['Date'])

intervalo_fecha_inicio = '2023-01-01'
intervalo_fecha_fin = '2023-09-01'
datos_filtrados = datos[(datos['Date'] >= intervalo_fecha_inicio) & (datos['Date'] <= intervalo_fecha_fin)]

plt.figure(figsize=(12, 6))
plt.plot(datos_filtrados['Date'], datos_filtrados['Open'], label='Precio de Inicio', color='red')
plt.plot(datos_filtrados['Date'], datos_filtrados['Close'], label='Precio de Cierre', color='blue')

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  # Cambia el intervalo de las fechas

plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Historia de Precios de Cierre e Inicio de Amazon (Ao 2023)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
